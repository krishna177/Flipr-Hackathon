```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
```

```python
data = pd.read_excel("Train_dataset.xlsx") 
data.head()
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
#StandardScaler()
#MinMaxScaler()

def scaleColumns(df, cols_to_scale):
    for col in cols_to_scale:
        df[col] = pd.DataFrame(min_max_scaler.fit_transform(pd.DataFrame(data[col])),columns=[col])
    return df
```

```python
def data_preprocessing(data):
    data=data.drop(columns=['Designation', 'Name'])
    data=data.fillna(0)
    data["Region"]= data["Region"].astype(str)
    data["Gender"]= data["Gender"].astype(str)
    data["Married"]= data["Married"].astype(str)
    data["Occupation"]= data["Occupation"].astype(str)
    data["Mode_transport"]= data["Mode_transport"].astype(str)
    data["comorbidity"]= data["comorbidity"].astype(str)
    data["cardiological pressure"]= data["cardiological pressure"].astype(str)
    class_lable = LabelEncoder()
    data['Married'] =class_lable.fit_transform(data['Married'].values)
    region = pd.get_dummies(data[['Region']])
    Gender = pd.get_dummies(data[['Gender']])
    Occupation = pd.get_dummies(data[['Occupation']])
    Mode_transport = pd.get_dummies(data[['Mode_transport']])
    cardiological_pressure = pd.get_dummies(data[['cardiological pressure']])
    comorbidity = pd.get_dummies(data[['comorbidity']])
    Pulmonary_score = pd.get_dummies(data[['Pulmonary score']])
    data=data.drop(columns=['Region', 'Gender','Occupation','Mode_transport','cardiological pressure','comorbidity','Pulmonary score'])
    df = scaleColumns(data,['people_ID','Age','Coma score','Platelets','Diuresis','Diuresis','Insurance','salary','HBB','d-dimer','Heart rate','HDL cholesterol','Charlson Index','Blood Glucose','cases/1M','Children','Deaths/1M'])
    df=pd.concat([df,region,Gender,Occupation,Mode_transport,cardiological_pressure,comorbidity,Pulmonary_score],axis=1)
    df=df.fillna(0)
    return df
```

```python
pros_df=data_preprocessing(data)
pros_df.head()
```

```python
y=pros_df["Infect_Prob"]
pros_df=pros_df.drop(columns=['Infect_Prob'])
y=y.div(100)
```

```python
pros_df.columns
```

```python
from numpy import sqrt
from pandas import read_csv
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras 
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Dropout
import numpy as np
```

```python
X_train, X_test, y_train, y_test = train_test_split(pros_df, y, test_size=0.20)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
n_features = X_train.shape[1]
```

```python
model = Sequential()
model.add(Dense(256, activation='sigmoid',kernel_initializer='random_normal', input_shape=(n_features,)))
model.add(Dropout(0.2))
model.add(Dense(124, activation='relu',kernel_initializer='random_normal'))
model.add(Dropout(0.5))
model.add(Dense(124, activation='relu',kernel_initializer='random_normal'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu',kernel_initializer='random_normal'))
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu',kernel_initializer='random_normal'))
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu',kernel_initializer='random_normal'))
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer="adam", loss="mse")
model.summary()
```

```python
model.fit(X_train, y_train, epochs=300,validation_data=(X_test, y_test),verbose=1)
```

```python
error = model.evaluate(X_test, y_test, verbose=0)
print('MSE: %.3f, RMSE: %.3f' % (error, sqrt(error)))
```
